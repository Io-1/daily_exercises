```python
def is_prime_base(n, m = 2, u10_primes = {2, 3, 5, 7}):
    
    #print(f"n = {n},  sqrt_n = {int(n**2)} with m = {m}") # check if base is prime was used

    if n in u10_primes: return True
    
    for l in range(m, int(n**0.5) + 1):
        if n % l == 0:
            return False
        
    if n in {0, 1}: return False
    
    return True
```

```python
def is_prime(n, primes):

    if n < 10: return is_prime_base(n)

    if (str(n)[-1] in {"0", "2", "4", "5", "6", "8"}): return False

    for p in primes:
        if n % p == 0: return False
        if p >= int(n**0.5): return True

    if not primes: primes = [2]
    
    return is_prime_base(n, m = primes[-1])
```

```python
def primorial_prep(tt = 18):
    primorial_primes, primorial,  mod_primorial_potential_set = [], [], []
    primes = {p for p in range(2, tt + 1) if is_prime_base(p)}
    for i, tt in enumerate(primes):
        primorial_primes.append({p for p in range(2, tt + 1) if is_prime_base(p)})
        primorial.append(reduce(operator.mul, primorial_primes[i], 1))
        mod_primorial_potential_set.append({j for j in range(primorial[i])}.difference({p * j for p in primorial_primes[i] for j in range(int(primorial[i] // p))}).union(primorial_primes[i]))
    return (primorial_primes, primorial, mod_primorial_potential_set)
```

```python
def is_prime_(n, primorial_primes, primorials, mod_primorial_potential_set):
    if n < 10: return is_prime_base(n)
    #if n != 49 else True
    #print(primorials)
    for primorial in primorials:
        print(primorial)
        if primorial == 2: continue
        if n % primorial not in mod_primorial_potential_set: return False
        for k in range(1, int(n**0.5) // primorial + 1):
            for i in mod_primorial_potential_set:
                if n % (k * primorial + i) == 0:
                    return False
        
        primes_to_primorial_max = {p for p in range(2, max(primorial_primes)) if is_prime_base(p)}

        for p in primes_to_primorial_max:
            if n % p == 0: return False

    if n == 4097: print(f"{n} will be prime because k was in range {int(n**0.5) // primorial + 1} with primorial {primorial} and primorial primes {primorial_primes}")
    return True
```
