```python
# https://leetcode.com/problems/insert-greatest-common-divisors-in-linked-list/

class Solution:
    def insertGreatestCommonDivisors(self, head: Optional[ListNode]) -> Optional[ListNode]:
        r = head
        tmp = head
        while tmp.next is not None:
            a = tmp.val
            b = tmp.next.val
            x = min(a, b)
            while True:
                if a % x == 0 and b % x == 0:
                    break
                x -= 1
            tmp = tmp.next
            r.next = ListNode(val=x, next=tmp)
            r = tmp
        return head
```

```python
# https://leetcode.com/problems/count-and-say/

class Solution:
    def countAndSay(self, n: int) -> str:
        if n > 1:
            rec = self.countAndSay(n - 1)
        elif n == 1:
            return "1"
        cnt = 0
        res = ""
        prev_d = rec[0]
        for d in rec:
            if d != prev_d:
                res += str(cnt) + prev_d
                cnt = 1
            else: cnt += 1
            prev_d = d
        res += str(cnt) + d
        return res
    
# A neat, non-recursion method

class Solution:
    def countAndSay(self, n: int) -> str:
        res = "1"
        for _ in range(n - 1):
            temp, i = "", 0
            while i < len(res):
                count = 1
                while i + 1 < len(res) and res[i] == res[i + 1]:
                    i += 1
                    count += 1
                temp += str(count) + res[i]
                i += 1
            res = temp
        return res
```

```python
# https://leetcode.com/problems/permutations/
class Solution:
    def _permutation(self, index: int, nums: list[int], permutation: list[int], res: list[list[int]], permutation_length: int) -> None:
        if len(permutation) == permutation_length:
            res.append(permutation.copy())
            
        for i, j in enumerate(nums[index:len(nums)] + nums[0:index]):
            permutation.append(j)
            self._permutation(0, nums[0:i] + nums[i + 1:len(nums)], permutation, res, permutation_length)
            if permutation: permutation.pop()

    def permute(self, nums: list[int]) -> list[list[int]]:
        res = []
        permutation_length = len(nums)
        self._permutation(0, nums, [], res, permutation_length)

        return res

# Improved solution removing time clutter, using backtrack permutating properly

class Solution:
    def permute(self, nums: list[int]) -> list[list[int]]:
        res = []

        def permutation(first: int) -> None:
            if first == len(nums):
                res.append(nums[:])
                return
            
            for i in range(first, len(nums)):
                nums[first], nums[i] = nums[i], nums[first]
                permutation(first + 1)
                nums[first], nums[i] = nums[i], nums[first]
        
        permutation(0)
        return res
```

```python
# https://leetcode.com/problems/combination-sum-ii/submissions/1716845081/

class Solution:
    def _combineForward(self, candidates: list[int], index: int, remaining_target: int, combination: List[int]):
        seen = set()
        
        if remaining_target == 0:
            self.res.append(combination.copy())
            print(f"success at: {combination}")
            return

        for i in range(index, len(candidates)):
            c = candidates[i]

            if c in seen: continue
            if remaining_target < c: break

            combination.append(c)
            self._combineForward(candidates, i + 1, remaining_target - c, combination)
            seen.add(c)
            combination.pop()

    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        self.res = []
        candidates.sort()

        self._combineForward(candidates, 0, target, [])
        return self.res        
```

```python
# https://leetcode.com/problems/combination-sum/

class Solution:
    def _combineForward(self, candidates: list[int], index: int, remaining_target: int, combination: List[int]):

        if remaining_target == 0:
            self.res.append(combination.copy())
            return

        for i in range(index, len(candidates)):
            c = candidates[i]

            if remaining_target < c: break

            combination.append(c)
            self._combineForward(candidates, i, remaining_target - c, combination)
            combination.pop()

    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        self.res = []
        candidates.sort()

        self._combineForward(candidates, 0, target, [])
        return self.res
```

```python
# https://leetcode.com/problems/swap-nodes-in-pairs/

class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next: return head

        neck = head.next

        neck.next = self.swapPairs(neck.next)
        head.next = neck.next
        neck.next = head
        
        return neck
```

```python
# https://leetcode.com/problems/container-with-most-water/

class Solution:
    def maxArea(self, height: List[int]) -> int:
        max_rec = 0
        i = 0
        j = len(height) - 1
        while i != j:
            rec = (j - i) * min(height[i], height[j])
            if rec > max_rec: max_rec = rec
            if height[i] < height[j]: i += 1
            else: j -= 1
        return max_rec
```

```python
# https://leetcode.com/problems/add-binary/

class Solution:
    def addBinary(self, a: str, b: str) -> str:
        a = list(a)
        b = list(b)
        mem = 0
        res = ""
        while a or b or mem:
            if a: x = int(a.pop())
            else: x = 0
            if b: y = int(b.pop())
            else: y = 0

            z = x + y + mem

            if z == 3:
                w = "1"
                mem = 1
            elif z == 2:
                w = "0"
                mem = 1
            elif z == 1:
                w = "1"
                mem = 0
            elif z == 0:
                w = "0"
                mem = 0
            res = w + res
        return res
```

```python
# https://leetcode.com/problems/reverse-integer/

class Solution:
    def reverse(self, x: int) -> int:

        int_min, int_max = -2**31, 2**31 - 1

        sgn = 1 if x >= 0 else -1
        x = abs(x)
        rev = 0

        while x != 0:
            pop = x % 10
            x = x // 10

            if rev > int_max // 10 or (rev == int_max // 10 and ((pop > int_max % 10 and sgn == 1) or (pop > int_min % 10 and sgn == -1))):
                return 0

            rev = rev * 10 + pop
        return sgn * rev
```

```python
# https://leetcode.com/problems/longest-palindromic-substring/

class Solution:
    def longestPalindrome(self, s: str) -> str:
        k, l = 0, 1
        idx_pairs = []
        for i in [2, 3]:
            for j in range(len(s) - i + 1):
                if s[j:j + i][-1] == s[j:j + i][0]: idx_pairs.append([j, j + i])
        while idx_pairs:
            k, l = idx_pairs.pop(0)
            if k != 0 and l != len(s) and s[k - 1] == s[l]:
                idx_pairs.append([k - 1, l + 1])
        return s[k:l]
```

```python
# https://leetcode.com/problems/zigzag-conversion/

class Solution:
    def convert(self, s: str, numRows: int) -> str:
        s_arr = list(s)
        result = ""
        lines = {i: "" for i in range(numRows)}
        while s_arr:
            for i in range(numRows):
                if s_arr:
                    lines[i] = lines[i] + s_arr.pop(0)
            for j in range(numRows - 2, 0, -1):
                if s_arr:
                    lines[j] = lines[j] + s_arr.pop(0)
        for l in lines.values():
            result = result + l
        return result
    
# Another neat solution:

class Solution:
    def convert(self, s: str, numRows: int) -> str:
        if numRows == 1 or numRows >= len(s):
            return s

        idx, d = 0, 1
        rows = [[] for _ in range(numRows)]

        for char in s:
            rows[idx].append(char)
            if idx == 0:
                d = 1
            elif idx == numRows - 1:
                d = -1
            idx += d

        for i in range(numRows):
            rows[i] = ''.join(rows[i])

        return ''.join(rows)   
```

```python
# https://leetcode.com/problems/median-of-two-sorted-arrays/description/

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        l1 = len(nums1)
        l2 = len(nums2)
        k = (l1 + l2 + 1) // 2
        for i in range(max(0, k - l2), min(k, l1) + 1):

            j = k - i

            n1_l = nums1[i - 1] if i != 0 else float("-inf")
            n1_r = nums1[i] if i != l1 else float('inf')
            n2_l = nums2[j - 1] if j != 0 else float("-inf")
            n2_r = nums2[j] if j != l2 else float('inf')

            if n1_l <= n2_r and n2_l <= n1_r:
                return max(n1_l, n2_l) if (l1 + l2) % 2 else (max(n1_l, n2_l) + min(n1_r, n2_r)) / 2
```

```python
# https://leetcode.com/problems/merge-two-sorted-lists/

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        head = None
        curr = ListNode(0)
        while list1 or list2:
            if list1 and list2:
                if list1.val <= list2.val:
                    curr.next = ListNode(list1.val)
                    curr = curr.next
                    if not head: head = curr
                    list1 = list1.next
                else:
                    curr.next = ListNode(list2.val)
                    curr = curr.next
                    if not head: head = curr
                    list2 = list2.next
            elif not list2:
                curr.next = ListNode(list1.val)
                curr = curr.next
                if not head: head = curr
                list1 = list1.next
            elif not list1:
                curr.next = ListNode(list2.val)
                curr = curr.next
                if not head: head = curr
                list2 = list2.next
        return head
    

            
# A way more compact solution using the l1, l2 lists themselves

class Solution:
    def mergeTwoLists(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        link = dummy

        while l1 and l2:
            if l1.val <= l2.val:
                link.next = l1
                l1 = l1.next
            else: 
                link.next = l2
                l2 = l2.next
            link = link.next

        link.next = l1 if l1 else l2

        return dummy.next
```

```python
# https://leetcode.com/problems/valid-parentheses/

class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        for b in s:
            if b in ("(", "[", "{"):
                stack.append(b)
            if b in (")", "]", "}"):
                if not stack: return False
                curr = stack.pop()
                if curr + b not in ("()", "[]", "{}"): return False
        if stack: return False
        return True
    
# Working a dictionary for comparing is a bit more comprehensible:

class Solution:
    def isValid(self, s: str) -> bool:
        if len(s) % 2:
            return False

        pairs = {')': '(', ']': '[', '}': '{'}
        stack = []

        for ch in s:
            if ch in pairs.values():        
                stack.append(ch)
            elif ch in pairs:               
                if not stack or stack.pop() != pairs[ch]:
                    return False
            else:
                return False

        return not stack
```

```python
# https://leetcode.com/problems/longest-common-prefix/

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        prefix = ""
        for i in range(len(min(strs, key = len))):
            ith_letters = {s[i] for s in strs}
            if len(ith_letters) > 1: return prefix
            prefix += ith_letters.pop()
        return prefix
```

```python
# https://leetcode.com/problems/roman-to-integer/

import re

class Solution:
    def romanToInt(self, s: str) -> int:
        map = {'M': 1000, 'CM': 900, 'D': 500, 'CD': 400, 'C': 100, 'XC': 90, 'L': 50, 'XL': 40, 'X': 10, 'IX': 9, 'V': 5, 'IV': 4, 'I': 1}
        r = 0
        for key in map:
            l_object = re.match(f"^(?:({key}))*", s)
            n = (len(l_object.group(0)) // len(key)) * map[key]
            r += n
            s = re.sub(f"^({key})*", "", s)
        return r
    
# But I could have just done it O(n) complexity instead of O(nk) where n is string length and k length of map

class Solution:
    def romanToInt(self, s: str) -> int:
        # 1) map char → valeur
        roman = {
            'I': 1,    'V': 5,
            'X': 10,   'L': 50,
            'C': 100,  'D': 500,
            'M': 1000
        }
        total = 0
        prev_value = 0

        # 2) on parcourt la chaîne à l'envers
        for ch in reversed(s):
            value = roman[ch]
            #  si un chiffre plus petit précède un plus grand, on soustrait
            if value < prev_value:
                total -= value
            else:
                total += value
            prev_value = value

        return total
```

```python
# https://leetcode.com/problems/add-two-numbers/

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        curr = dummy
        carry = 0


        while l1 or l2 or carry:
            n1, n2 = l1.val if l1 else 0, l2.val if l2 else 0
            s = n1 + n2 + carry
            carry = s // 10

            curr.next = ListNode(s % 10)
            curr = curr.next

            if l1: l1 = l1.next
            if l2: l2 = l2.next

        return dummy.next
```
