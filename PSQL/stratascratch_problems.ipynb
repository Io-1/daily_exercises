```python
# https://platform.stratascratch.com/coding/2110-salary-less-than-twice-the-average

"""
WITH empl_avg_cte AS (
    SELECT m.manager_empl_id, AVG(e.salary) AS avg_employee_salary
    FROM map_employee_hierarchy m
        JOIN dim_employee e
            ON m.empl_id = e.empl_id
    GROUP BY m.manager_empl_id
)

SELECT c.manager_empl_id, me.salary as manager_salary, c.avg_employee_salary
FROM empl_avg_cte c
    JOIN dim_employee me
        ON c.manager_empl_id = me.empl_id
WHERE me.salary < 2 * c.avg_employee_salary
"""
```

```python
# https://platform.stratascratch.com/coding/2056-number-of-shipments-per-month

"""
WITH month_cte AS (
    SELECT shipment_id, sub_id, to_char(shipment_date, 'YYYY-MM') AS year_month
    FROM amazon_shipment
)

SELECT year_month, COUNT(*)
FROM month_cte
GROUP BY year_month;
"""
```

```python
# https://platform.stratascratch.com/coding/2024-unique-users-per-client-per-month

"""
WITH timed_cte AS (
    SELECT client_id, user_id, EXTRACT(MONTH FROM time_id) AS month_id
    FROM fact_events
)

SELECT client_id, month_id, COUNT(DISTINCT user_id)
FROM timed_cte
GROUP BY client_id, month_id;
"""
```
